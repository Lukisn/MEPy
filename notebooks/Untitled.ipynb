```python
source_path = "../kostra/original/Raster_geog_bezug.csv"
target_path = "../kostra/raster_clean.csv"

def clean_csv_file(source_file, target_file):
    with open(source_path, "r") as source_file:
        with open(target_path, "w") as target_file:
            for line in source_file:
                target_file.write(line.replace(",", ".").replace(";", ","))
                
clean_csv_file(source_path, target_path)
```
